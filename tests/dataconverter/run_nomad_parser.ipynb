```python
# pip install pynxtools in editable mode mem_optimization_parsing branch, nomad-lab, flatdict
try:
    from nomad.datamodel import EntryArchive
    from nomad.units import ureg
    from nomad.utils import get_logger
except ImportError:
    pytest.skip("nomad not installed", allow_module_level=True)

from pynxtools.nomad.parsers.parser import NexusParser
import flatdict as fd
import time
# import tracemalloc

prefix = "mem_opt_nomad_parse"
```

```python
n = 2_500_000_000
for compress in [True, False]:
    for storage_layout in ["contiguous", "chunked"]:
        if storage_layout == "contiguous":
            suffix = ""
            if not compress:
                continue
        else:
            suffix = f"_{'compressed' if compress else 'uncompressed'}"
        file_path = f"{prefix}_{n}_{storage_layout}{suffix}.nxs"
        
        print(f">>>> {file_path}")
        # continue
        
        tic = time.perf_counter()  # microsecond  _ns()
      
        archive = EntryArchive()
        NexusParser().parse(file_path, archive, get_logger(__name__))
        # em_obj = archive.data
    
        toc = time.perf_counter()
        print(f"took {(toc - tic)} s")  #  / 1.e9} s")
        continue
        
        for key, obj in fd.FlatDict(archive.m_to_dict(), delimiter="/").items():
            # if key == "results":
            print(f"{key}, {obj}")
```
