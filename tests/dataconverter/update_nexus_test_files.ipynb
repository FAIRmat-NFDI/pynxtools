```python
import os
from pynxtools_em.parsers.hfive_base import HdfFiveBaseParser
print(os.getcwd())
import h5py
import shutil
from jupyterlab_h5web import H5Web
```

```python
for nxsname in ["contiguous", "chunked_uncompressed", "chunked_compressed"]:
    in_file_path = f"mem_opt_nomad_parse_{2_500_000}_{nxsname}.nxs"
    ou_file_path = f"{in_file_path}.sha256.test.yaml"
    
    p = HdfFiveBaseParser(in_file_path)
    p.get_content()
    p.store_hashes(blacklist_by_key=(), blacklist_by_suffix=(), file_path=ou_file_path)
```

```python
ref_tmp = [
    ("../ellips/pynxtools-ellips/tests/data", "test_ref.nxs"),
    ("/tmp/pytest-of-kaiobach/pytest-0/test_nexus_conversion_reader_t0", "output.nxs"),
]
for dirname, nxsname in ref_tmp:
    print(f"{dirname}, {nxsname}")
    in_file_path = f"{dirname}/{nxsname}"
    ou_file_path = f"{dirname}/{nxsname}.sha256.test.yaml"

    p = HdfFiveBaseParser(in_file_path)
    p.get_content()
    p.store_hashes(blacklist_by_key=(), blacklist_by_suffix=(), file_path=ou_file_path)

    if nxsname == "output.nxs":
        shutil.move(ou_file_path, f"{ref_tmp[0][0]}/{ou_file_path[ou_file_path.rfind('/') + 1:]}")
```
