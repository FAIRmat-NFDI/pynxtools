```python
from contextlib import redirect_stdout, redirect_stderr
from pynxtools.dataconverter.validate_file import validate

import time
from numpy import arange
# prefix = "stack3d_when_it_should_be_stack2d"
prefix = "mem_opt_nomad_track"
```

```python
n = 2_500_000  # _000
for compress in [True, False]:
    for storage_layout in ["contiguous", "chunked"]:
        if storage_layout == "contiguous":
            suffix = ""
            if not compress:
                continue
        else:
            continue
            suffix = f"_{'compressed' if compress else 'uncompressed'}"
            
        file_path = f"{prefix}_{n}_{storage_layout}{suffix}.nxs"
        print(f">>>> {file_path}")

        for idx in arange(10):
            with open(f"stdout.{idx}.log", "w") as out:  #, open(f"stderr.{idx}.log", "w") as err:
                with redirect_stdout(out):  #, redirect_stderr(err):
                    validate(f"{file_path}")
        # tic = time.perf_counter()  # microsecond  _ns()
        # toc = time.perf_counter()
        # print(f"took {(toc - tic)} s")  #  / 1.e9} s")
```
